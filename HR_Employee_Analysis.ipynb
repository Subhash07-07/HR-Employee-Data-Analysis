```python
df.drop(columns=["EmployeeCount", "Over18", "StandardHours"], inplace=True, errors='ignore')
df.shape
df.columns
```

```python
df.nunique().sort_values()
```

```python
#Age group
df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[17, 25, 35, 45, 60],
    labels=["18-25", "26-35", "36-45", "46-60"]
)
df[["Age", "Age_Group"]].head()
```

```python
#Income Category
df["Income_Category"] = pd.cut(
    df["MonthlyIncome"],
    bins=[0, 5000, 10000, 15000, df["MonthlyIncome"].max()],
    labels=["Low", "Medium", "High", "Very High"]
)
df[["MonthlyIncome", "Income_Category"]].head()
```

```python
#Experience Level
df["Experience_Level"] = pd.cut(
    df["TotalWorkingYears"],
    bins=[-1, 5, 10, 20, 40],
    labels=["Beginner", "Intermediate", "Experienced", "Expert"]
)
df[["TotalWorkingYears", "Experience_Level"]].head()
```

```python
#What is the Overall Attrition Rate?
df["Attrition"].value_counts()
df["Attrition"].value_counts(normalize=True) * 100
```

```python
# Gender Distribution
df["Gender"].value_counts()
df["Gender"].value_counts(normalize=True) * 100
```

```python
# Department-wise Employee Count
df["Department"].value_counts()
```

```python
# Average Monthly Income by Department
df.groupby("Department")["MonthlyIncome"].mean().sort_values(ascending=False)
```

```python
# : Average Income by Job Role
df.groupby("JobRole")["MonthlyIncome"].mean().sort_values(ascending=False)
```

```python
# Attrition by Department
pd.crosstab(df["Department"], df["Attrition"])
pd.crosstab(
    df["Department"],
    df["Attrition"],
    normalize="index"
) * 100
```

```python
# Overtime vs Attrition
pd.crosstab(df["OverTime"], df["Attrition"])
pd.crosstab(
    df["OverTime"],
    df["Attrition"],
    normalize="index"
) * 100
```

```python
# Average Income by Education Field
df.groupby("EducationField")["MonthlyIncome"].mean().sort_values(ascending=False)
```

```python
# Average Monthly Income by Age Group
df.groupby("Age_Group")["MonthlyIncome"].mean()
```

```python
# Performance Rating Distribution
df["PerformanceRating"].value_counts()
```

```python
# Attrition Distribution
df["Attrition"].value_counts().plot(
    kind="bar",
    figsize=(6,4),
    title="Employee Attrition"
)

plt.xlabel("Attrition")
plt.ylabel("Number of Employees")
plt.show()
```

```python
# Gender Distribution
df["Gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.ylabel("")
plt.title("Gender Distribution")
plt.show()
```

```python
# Employees by Department
df["Department"].value_counts().plot(
    kind="bar",
    figsize=(8,5),
    title="Employees by Department"
)

plt.xlabel("Department")
plt.ylabel("Employees")
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Monthly Income by Department
df.groupby("Department")["MonthlyIncome"] \
  .mean() \
  .sort_values() \
  .plot(
      kind="barh",
      figsize=(8,5),
      title="Average Monthly Income by Department"
)

plt.xlabel("Average Monthly Income")
plt.show()
```

```python
 # Average Monthly Income by Job Role
df.groupby("JobRole")["MonthlyIncome"] \
  .mean() \
  .sort_values() \
  .plot(
      kind="barh",
      figsize=(9,6),
      title="Average Monthly Income by Job Role"
)

plt.xlabel("Average Monthly Income")
plt.show()
```

```python
# Average Monthly Income by Job Role
df.groupby("JobRole")["MonthlyIncome"] \
  .mean() \
  .sort_values() \
  .plot(
      kind="barh",
      figsize=(9,6),
      title="Average Monthly Income by Job Role"
)

plt.xlabel("Average Monthly Income")
plt.show()
```

```python
# Overtime Distribution
df["OverTime"].value_counts().plot(
    kind="bar",
    figsize=(6,4),
    title="Overtime Distribution"
)

plt.show()
```
